# Import

```python
import cv2
import numpy as np
```

# Convert

```python
def ConvertRGD(img_path, depth_path, out_path):
  directory_contents = os.listdir(img_path)
  existing_contents = os.listdir(out_path)

  for item in directory_contents:
    img_filename = img_path + "/" + item
    depth_filename = depth_path + "/" + item

    if not os.path.isdir(img_filename):
      if item in existing_contents:
        print(item + " - converted to RGD, moving on...")
        continue

      # in BGR format
      img = cv2.imread(img_filename)
      depth = cv2.imread(depth_filename)

      r = img[:,:,2]
      g = img[:,:,1]
      d = np.expand_dims(depth[:,:,0], axis=2)

      rgd = np.dstack((d, g, r))

      #print(r.shape)
      #print(g.shape)
      #print(d.shape)
      #print(rgd.shape)

      cv2.imwrite(out_path + "/" + item, rgd)
```

```python
def ConvertRDB(img_path, depth_path, out_path):
  directory_contents = os.listdir(img_path)
  existing_contents = os.listdir(out_path)

  for item in directory_contents:
    img_filename = img_path + "/" + item
    depth_filename = depth_path + "/" + item

    if not os.path.isdir(img_filename):
      if item in existing_contents:
        print(item + " - converted to RDB, moving on...")
        continue

      img = cv2.imread(img_filename)
      depth = cv2.imread(depth_filename)

      r = img[:,:,0]
      b = np.expand_dims(img[:,:,2], axis=2)
      d = depth[:,:,0]

      rdb = np.dstack((b, d, r))

      cv2.imwrite(out_path + "/" + item, rdb)
```

```python
def ConvertDGB(img_path, depth_path, out_path):
  directory_contents = os.listdir(img_path)
  existing_contents = os.listdir(out_path)

  for item in directory_contents:
    img_filename = img_path + "/" + item
    depth_filename = depth_path + "/" + item

    if not os.path.isdir(img_filename):
      if item in existing_contents:
        print(item + " - converted to DGB, moving on...")
        continue

      img = cv2.imread(img_filename)
      depth = cv2.imread(depth_filename)

      d = depth[:,:,0]
      g = img[:,:,1]
      b = np.expand_dims(img[:,:,2], axis=2)
      dbg = np.dstack((b, g, d))

      cv2.imwrite(out_path + "/" + item, dbg)
```

```python
#ConvertRGD('coco_validate/data', 'coco_validate/depth', 'coco_validate/rgd')
#ConvertRDB('coco_validate/data', 'coco_validate/depth', 'coco_validate/rdb')
ConvertDGB('coco_validate/data', 'coco_validate/depth', 'coco_validate/dgb')
```

```python
#ConvertRGD('coco_train/data', 'coco_train/depth', 'coco_train/rgd')
#ConvertRDB('coco_train/data', 'coco_train/depth', 'coco_train/rdb')
ConvertDGB('coco_train/data', 'coco_train/depth', 'coco_train/dgb')
```
