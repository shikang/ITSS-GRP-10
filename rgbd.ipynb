#Combine RGB & D

```python
def CombineRGBD(img_path, depth_path, out_path):
  directory_contents = os.listdir(img_path)
  existing_contents = os.listdir(out_path)

  for item in directory_contents:
    img_filename = img_path + "/" + item
    depth_filename = depth_path + "/" + item

    if not os.path.isdir(img_filename):
      out_item = os.path.splitext(item)[0]+'.png'
      out_filename = out_path + "/" + out_item
      #print(out_item)
      print(out_filename)

      if out_item in existing_contents:
        print(out_item + " - converted to RGBD, moving on...")
        continue

      img = Image.open(img_filename).convert('RGB')
      depth = Image.open(depth_filename).convert('L')

      img.putalpha(depth)

      img.save(out_filename)
```

```python
CombineRGBD('coco_test/data', 'coco_test/kitti-depth', 'coco_test/kitti-rgbd')
CombineRGBD('coco_train/data', 'coco_train/kitti-depth', 'coco_train/kitti-rgbd')

#CombineRGBD('coco_validate/data', 'coco_validate/depth', 'coco_validate/midas-rgbd')
```
